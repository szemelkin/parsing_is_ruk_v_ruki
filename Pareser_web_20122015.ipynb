```python
def web_parse(url_for_web_parsing):
    import time
    #Парсим страницу каждого объявления
    s2 = BeautifulSoup(requests.get(url_for_web_parsing).text, "html5lib")
    ads_name=s2.findAll('div', {'class' : 'advertOwnerEntry'})
    #print ads_name[0].text.strip()

    ads_description=s2.findAll('div', {'class' : 'advertDescriptionText'})
    #print ads_description[0].text.strip()

    ads_price=s2.findAll('div', {'class' : 'productPagePrice'})
    #print ads_price[0].text.strip()

    ads_productname=s2.findAll('', {'class' : 'productPageName'})
    #print ads_productname[0].text.strip()

    ads_createdate=s2.findAll('div', {'class' : 'createDate'})
    #print ads_createdate[0].text.strip()

    #Формируем строку для записи в файл
    string_to_file = ''
    try:
        string_to_file = ads_name[0].text.strip().encode('cp1251')+';'
    except: string_to_file = 'нет имени или города'+';'
    try:
        string_to_file = string_to_file+ads_name[1].text.strip().encode('cp1251')+';'
    except: string_to_file =string_to_file + 'нет имени или города'+';'
    try:
        string_to_file = string_to_file+ads_description[0].text.strip().encode('cp1251')+';'
    except: string_to_file =string_to_file + 'нет описания'+';'
    try:
        string_to_file = string_to_file+ads_price[0].text.strip().encode('cp1251')+';'
    except: string_to_file =string_to_file + 'нет цены'+';'
    try:
        string_to_file = string_to_file+url_for_detailed_info[0].strip().encode('cp1251')+';'
    except: string_to_file =string_to_file + 'нет деталей'+';'
    try:
        string_to_file = string_to_file+ads_productname[0].text.strip().encode('cp1251')+';'
    except: string_to_file =string_to_file + 'нет названия'+';'  
    try:
        string_to_file = string_to_file+ads_createdate[0].text.strip().encode('cp1251')
    except: string_to_file =string_to_file + 'нет даты создания'+';'
    time.sleep(0)
    return string_to_file

def write_to_file(str_to_file):
    text_file = open("Irvr_parse.txt", "a")
    text_file.write(str_to_file + '\n')
    text_file.close
```
